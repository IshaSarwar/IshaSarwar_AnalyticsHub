## Heirarchical Clustring
- Hierarchical clustring is a method for grouping data into disjoint subsets of similar data.
- It is a non-parametric method that is used when the number of clusters is not known in advance.
- It is also used to find clusters in data that is not normally distributed.
- The method is based on the idea that similar data points are likely to be grouped together.
- The method starts with a single cluster containing all data points, and then iteratively splits the cluster into
two sub-clusters until a stopping criterion is met.
- The stopping criterion can be based on the number of clusters, the size of the clusters, or the similarity between the clusters.
- The hierarchical clustring method can be implemented using a variety of algorithms, including the
single-linkage, complete-linkage and average-linkage methods.




```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs
```

```python
# Generate a random dataset with 100 samples and 3 features
X, y = make_blobs(n_samples=100, centers=11, n_features=3, random_state=42)

# Perfrom hierarchical clustring with the agglomerative clustring algorithem
clustring = AgglomerativeClustering(n_clusters=3,linkage='ward')
# Fit the model to the data
clustring.fit(X)

# Create a dendrogram to visualize the hierarchical clustering
plt.figure(figsize=(10, 5))
plt.title('Dendogram for Hierarchical Clustring')
dendogram = sch.dendrogram(sch.linkage(X, method='ward'))

plt.show()
plt.show()
```

```python
# load the dataset
df= sns.load_dataset('tips')

df.head()

```

```python
# Select relevant features for clustering
# Here, we will use 'total_bill', 'size', and 'tip' as features
X =df[['total_bill','size', 'tip']]

#perform hierarchical clustring
clustering_tips = AgglomerativeClustering(n_clusters=5 , linkage='ward')
clustering_tips.fit(X)
# Add the cluster labels to the original DataFrame
df['cluster_tips'] = clustering_tips.labels_
# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.title('Hierarchical Clustring for Tips Dataset')
sns.scatterplot(data=df, x='total_bill', y='tip', hue='cluster_tips', palette='Set2')

plt.show()



```
